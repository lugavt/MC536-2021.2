## Tarefa 1

### Carregue o nó de Tenda dos Milagres (da versão em inglês Tent of Miracles) em Python e mostre o conteúdo JSON.

```python
import io
import requests
import json
```

```python
urlMiracles = 'https://dbpedia.org/data/Tent_of_Miracles_(novel).jsod'
dataMiracles = requests.get(urlMiracles)
jsonMiracles = dataMiracles.json()
print(json.dumps(jsonMiracles, indent = 3))
```

## Tarefa 2
### Recupere via Python e mostre o autor de Tenda dos Milagres?

```python
autor = jsonMiracles["d"]["results"][0]["http://dbpedia.org/property/author"]["__deferred"]["uri"]
autor = autor[str.rindex(autor,"/")+1:]
autor = autor.replace("_"," ")
print(autor)
```

## Tarefa 3
### Quais os termos usados para Aspirina (consulta JSON/MeSH em Python)?

```python
url = "https://id.nlm.nih.gov/mesh/lookup/descriptor?label=Aspirin&match=exact&limit=1"
data = requests.get(url)
jsonAspirin = data.json()
print(json.dumps(jsonAspirin, indent=3))

#para extrair o codigo da Aspirina
resource = jsonAspirin[0]["resource"]
codigo = resource[str.rindex(resource,"/")+1 : ]
print(codigo)
```

```python
url = "https://id.nlm.nih.gov/mesh/lookup/details?descriptor={codigo}"
url = url.replace("{codigo}", codigo)
data = requests.get(url)
jsonAspirin = data.json()
terms = []
for i in range(len(jsonAspirin["terms"])):
    terms.append(jsonAspirin["terms"][i]["label"])
    print(terms[i])
```

## Tarefa 4
### Escolha algum lugar no OpenStreetMap e elabore a mesma sequência de três chamadas de serviços para o seu conceito.

```python
url_template = "http://nominatim.openstreetmap.org/search?format={format}&q={query}"
format = "json"
query = "Stanford University"

url = url_template.replace("{format}", format)\
                  .replace("{query}", query)

data = requests.get(url)
stanford  = data.json()
print(json.dumps(stanford, indent=3))
```

```python
url_template = "https://api.openstreetmap.org/api/0.6/{element_type}/{id}.{format}"

element_type = "way" # types: node, way, or relation
id = str(stanford[0]["osm_id"])
format = "json"

url = url_template.replace("{element_type}", element_type)\
                  .replace("{id}", id)\
                  .replace("{format}", format)

data = requests.get(url)
stanfordWay = data.json()
print(json.dumps(stanfordWay, indent=3))
```

```python
url_template = "http://dbpedia.org/data/{concept}.{format}"
concept = stanfordWay["elements"][0]["tags"]["wikipedia"].replace(" ", "_")\
                                                        .replace("en:", "")
format = "json"

url = url_template.replace("{concept}", concept)\
                  .replace("{format}", format)

data = requests.get(url)
stanford_dbpedia = data.json()
print(json.dumps(stanford_dbpedia, indent=3))
```
